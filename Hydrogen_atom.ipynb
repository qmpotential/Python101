```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def hydrogen_1s(x, y, z):
    # Function for the 1s orbital wavefunction
    r = np.sqrt(x**2 + y**2 + z**2)
    return (1/np.sqrt(np.pi)) * np.exp(-r)

# Define the grid of points in Cartesian coordinates
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
z = np.linspace(-5, 5, 100)
X, Y, Z = np.meshgrid(x, y, z)

# Calculate the wavefunction values for each point
wavefunction = hydrogen_1s(X, Y, Z)

# Plot the 1s orbital
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('1s Orbital')

# Create the surface plot
surf = ax.contourf(X, Y, Z, wavefunction, cmap='hot')
fig.colorbar(surf)

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def hydrogen_1s(r, R):
    # Function for the 1s orbital wavefunction
    return (1/np.sqrt(np.pi)) * np.exp(-2 * np.abs(r - R))

# Define the radial distance range
r = np.linspace(0, 10, 100)

# Set the equilibrium bond length
R = 0.74

# Calculate the wavefunction values for each radial distance
wavefunction1 = hydrogen_1s(r, 0)  # Orbital centered at r=0
wavefunction2 = hydrogen_1s(r, R)  # Orbital centered at r=R

# Plot the orbitals
plt.plot(r, wavefunction1, label='Hydrogen Orbital 1')
plt.plot(r, wavefunction2, label='Hydrogen Orbital 2')
plt.xlabel('Radial Distance')
plt.ylabel('Wavefunction')
plt.title('Hydrogen Orbitals at Equilibrium Position')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def hydrogen_1s(r, R):
    # Function for the 1s orbital wavefunction
    return (1/np.sqrt(np.pi)) * np.exp(-2 * np.sqrt(r**2 + R**2))

# Define the radial distance and angular coordinates
r = np.linspace(-5, 5, 100)
theta = np.linspace(0, np.pi, 100)
phi = np.linspace(0, 2*np.pi, 100)

# Create the grid of points in spherical coordinates
R, Theta, Phi = np.meshgrid(r, theta, phi)

# Convert spherical coordinates to Cartesian coordinates
X = R * np.sin(Theta) * np.cos(Phi)
Y = R * np.sin(Theta) * np.sin(Phi)
Z = R * np.cos(Theta)

# Set the equilibrium bond length
bond_length = 0.74

# Calculate the wavefunction values for each point
wavefunction1 = hydrogen_1s(np.sqrt(X**2 + Y**2 + Z**2), 0)  # Orbital centered at (0, 0, 0)
wavefunction2 = hydrogen_1s(np.sqrt((X-bond_length)**2 + Y**2 + Z**2), bond_length)  # Orbital centered at (bond_length, 0, 0)

# Plot the orbitals
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Hydrogen Orbitals at Equilibrium Position')

# Create the surface plot for the first orbital
surf1 = ax.plot_surface(X, Y, Z, cmap='hot', alpha=0.7)
surf1.set_array(wavefunction1.flatten())

# Create the surface plot for the second orbital
surf2 = ax.plot_surface(X-bond_length, Y, Z, cmap='cool', alpha=0.7)
surf2.set_array(wavefunction2.flatten())

# Add a colorbar
fig.colorbar(surf1, ax=ax, label='Wavefunction')

# Display the plot
plt.show()
```
